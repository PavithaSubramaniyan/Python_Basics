**#Difference between break, continue , pass**

*   Break terminates loop immediately
*   Continue skip the current iteration and moves to next one


*   Pass acts as nothing it is just a placeholder when the statement is syntactically required but no action is needed







```
for i in range(5):
  if i == 2:
    break
  elif i ==1:
    continue
  else:
    pass
  print(i) #skips 1 and break at 2
```

```
#count divisiors of a number
def divisors(num):
  lis = []
  for i in range(1,num+1):
    if num%i == 0:
      lis.append(i)
  return lis

print(divisors(int(input())))
```

**5. MRO or Diamond Pattern**



*   Method Resolution defines the class search order in Hybrid inheritance
*   Resolving diamond problem via c3 linearization





```
class A:
  def show(self): print("A")
class B(A):
   def show(self): print("B")
class C(A):
  def show(self): print("C")
class D(B, C): pass

d = D()
d.show()  # B (MRO: D->B->C->A)
print(D.mro())  # [<class '__main__.D'>, b,c,a]
```

```
#Two string count vowels and compare length
def vowel_len(str1, str2):
  count1 = 0
  vowel = 'aeiou'
  for i in str1:
    if i in vowel:
      count1 += 1
  count2 = [i for i in str2 if i in vowel]
  if count1 == len(count2):
    return "Same length"
  else:
    return "vowels not same length"
str1 = "hello"
str2 = "world"
print(vowel_len(str1, str2))
```

**9. Map vs Filter**


*   Map applies function to every iterable item and return transformed results
*   Filter applies function returning only items where it evaluates True

```
a = [1,2,3,4,5]
print(list(map(lambda x:x*x, a)))
# filter
print(list(filter(lambda x : x%2 == 0, a)))
```

```
#Remove duplicates from string
def remove_dup(str):
  temp=''
  for i in str:
    if i not in temp:
      temp += i
  return temp
print(remove_dup(input()))
```

```
#most occuring character(s)
from collections import Counter
def most_ocuur(str):
  c = Counter(str)
  maxval = max(c.values())
  val = []
  for k, v in c.items():
    if maxval == v:
      val.append(k)
  return val
print(most_ocuur(input()))
```

**Multiple inheritance example**


*   Single child inherits from multiple parents
*   MRO resolves method order



```
class A:
  def show(self): print("A")
class B:
  def show(self): print("B")
class C(A,B): pass
c = C()
c.show()
```

**Generator example**

*   Uses yield for lazy iteration
*   memory efficient



```
def gen_square(n):
  for i in range(n):
    yield i**2
print(list(gen_square(3)))
```
